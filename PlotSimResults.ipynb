### Import Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

sim_data = pd.read_csv("run_sim_all.csv",index_col=0)
sim_data
```

```python
def plt_for_rep_types(plt_lines, x_type, log_x=True):
    # Set up the axes
    gs = plt.GridSpec(1,3)
    fig = plt.figure(figsize=(21,5))
    ax1 = plt.subplot(gs[0,0])
    
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax2 = plt.subplot(gs[0,1])
    
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax3 = plt.subplot(gs[0,2])
    
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    
    mean_vals = sim_data.groupby([plt_lines, 'Replacement Type', x_type]).mean()
    vals = sim_data[plt_lines].unique()
    vals.sort()
    for assoc in vals:
        ax1.plot(mean_vals.loc[assoc, "LRU"].index, mean_vals.loc[assoc, "LRU"]["Cache Miss Rate (%)"].values)
        ax2.plot(mean_vals.loc[assoc, "RR"].index, mean_vals.loc[assoc, "RR"]["Cache Miss Rate (%)"].values )
        ax3.plot(mean_vals.loc[assoc, "RND"].index, mean_vals.loc[assoc, "RND"]["Cache Miss Rate (%)"].values)
    
    if log_x:    
        ax1.set_xscale('log')
        ax2.set_xscale('log')
        ax3.set_xscale('log')

    ax2.set_xlabel(x_type, fontsize=20)
    ax1.set_ylabel("Cache Miss Rate (%)", fontsize=20)
    ax1.set_title("LRU", fontsize=20)
    ax2.set_title("RR", fontsize=20)
    ax3.set_title("RND", fontsize=20)

    legend_txt = []
    for l in vals:
        legend_txt += [plt_lines + " " + str(l)]
    plt.legend(legend_txt, bbox_to_anchor=(1.4,1.0))

    plt.savefig(os.path.join("images", plt_lines + '-' + x_type + '_lineplot.jpg'))
    plt.show()
```

```python
plt_for_rep_types("Block Size", "Cache Size")
```

```python
plt_for_rep_types('Associativity', "Cache Size")
```

```python
def do_hit_mis_bars(ind):
    # Set up the axes
    gs = plt.GridSpec(1,3)
    fig = plt.figure(figsize=(30,5))
    ax1 = plt.subplot(gs[0,0])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax2 = plt.subplot(gs[0,1])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax3 = plt.subplot(gs[0,2])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")

    mean_vals = sim_data.groupby(['Replacement Type', ind]).mean()
    index_ints_lru = mean_vals.loc["LRU"].index
    index_ints_rr = mean_vals.loc["RR"].index
    index_ints_rnd = mean_vals.loc["RND"].index
    lru_idx = ["%d" % d for d in index_ints_lru]
    rr_idx = ["%d" % d for d in index_ints_rr]
    rnd_idx = ["%d" % d for d in index_ints_rnd]
    
    #m1 = [mean_vals.loc["LRU"]["Cache Hits"].values,
    #      mean_vals.loc["RR"]["Cache Hits"].values, 
    #      mean_vals.loc["RND"]["Cache Hits"].values]
    
    #m2 = [mean_vals.loc["LRU"]["Cache Misses"].values, 
    #      mean_vals.loc["RR"]["Cache Misses"].values, 
    #      mean_vals.loc["RND"]["Cache Misses"].values]
    
    m3 = [mean_vals.loc["LRU"]["Compulsory Cache Misses"].values,
          mean_vals.loc["RR"]["Compulsory Cache Misses"].values, 
          mean_vals.loc["RND"]["Compulsory Cache Misses"].values]
    
    m4 = [mean_vals.loc["LRU"]["Conflict Misses"].values,
          mean_vals.loc["RR"]["Conflict Misses"].values, 
          mean_vals.loc["RND"]["Conflict Misses"].values]
    
    #ax1.bar(np.arange(0,len(lru_idx)), m1[0], ec='k')
    #ax2.bar(np.arange(0,len(rr_idx)), m1[1], ec='k')
    #ax3.bar(np.arange(0,len(rnd_idx)), m1[2], ec='k')
    
    #ax1.bar(np.arange(0,len(lru_idx)), m2[0], bottom=m1[0], ec='k')
    #ax2.bar(np.arange(0,len(rr_idx)), m2[1], bottom=m1[1], ec='k')
    #ax3.bar(np.arange(0,len(rnd_idx)), m2[2], bottom=m1[2], ec='k')

    ax1.bar(np.arange(0,len(lru_idx)), m3[0], bottom=0, ec='k', tick_label=lru_idx)
    ax2.bar(np.arange(0,len(rr_idx)), m3[1], bottom=0, ec='k', tick_label=rr_idx)
    ax3.bar(np.arange(0,len(rnd_idx)), m3[2], bottom=0, ec='k', tick_label=rnd_idx)
    
    ax1.bar(np.arange(0,len(lru_idx)), m4[0], bottom=m3[0], ec='k', color='C3', tick_label=lru_idx)
    ax2.bar(np.arange(0,len(rr_idx)), m4[1], bottom=m3[1], ec='k', color='C3', tick_label=rr_idx)
    ax3.bar(np.arange(0,len(rnd_idx)), m4[2], bottom=m3[2], ec='k', color='C3', tick_label=rnd_idx)

    ax2.set_xlabel(ind, fontsize=20)
    ax1.set_ylabel("Number of occurrences", fontsize=20)
    ax1.set_title("LRU", fontsize=20)
    ax2.set_title("RR", fontsize=20)
    ax3.set_title("RND", fontsize=20)
    
    plt.legend(["Compulsory Cache Misses", "Conflict Misses"], bbox_to_anchor=(1.4,1.0))
    plt.savefig(os.path.join("images", ind + '_barplot.jpg'))
    plt.show()
```

```python
def plot_on_rep_tp():
    # Set up the axes
    gs = plt.GridSpec(1,4)
    fig = plt.figure(figsize=(30,5))
    ax1 = plt.subplot(gs[0,0])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax2 = plt.subplot(gs[0,1])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax3 = plt.subplot(gs[0,2])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")
    ax4 = plt.subplot(gs[0,3])
    plt.rc('axes', axisbelow=True)
    plt.grid(linestyle=":")

    mean_vals = sim_data.groupby(['Replacement Type']).mean()
    
    ax1.bar(["LRU", "RR", "RND"], mean_vals["Compulsory Cache Misses"].values, ec='k')
    ax2.bar(["LRU", "RR", "RND"], mean_vals["Conflict Misses"].values, ec='k')
    ax3.bar(["LRU", "RR", "RND"], mean_vals["Cache Hits"].values, ec='k')    
    ax4.bar(["LRU", "RR", "RND"], mean_vals["Cache Miss Rate (%)"].values, ec='k')    
    
    ax1.set_ylabel("Compulsory Cache Misses")
    ax2.set_ylabel("Conflict Misses")
    ax3.set_ylabel("Cache Hits")
    ax4.set_ylabel("Cache Miss Rate (%)")
    plt.savefig(os.path.join("images", "rep_cachemiss_barplot.jpg"))
    plt.show()
plot_on_rep_tp()
```
